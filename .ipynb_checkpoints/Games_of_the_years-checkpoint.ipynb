```python
def get_award(line):
    awards = []
    for i in line:
        if str(type(i)) == "<class 'bs4.element.Tag'>":
            #print(i.get_text())
            awa = i.get_text().strip("\n").split("\n")
            for y in range(awa.count("")):
                awa.remove("")
            awards.append(awa)
    return awards
```

```python
def table_dic(table):
    dics = []
    n = 0
    columns = [re.sub("\[\d]", "", i.get_text()[:-1]) for i in table.find_all("th")]
    for i in table:
        line = get_award(i)
        for y in line[1:]:
            dics.append({})
            for x in range(len(columns)):
                dics[n][columns[x]] = y[x]
            dics[n][columns[x]] = y[x:]
            n +=1
    return dics
```

```python
lst = []
for i in awards:
    lst.append(table_dic(i))
```

```python
def get_tables(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    return soup.find_all("table", class_="wikitable")
```

```python
def to_dic(table):
    values = []
    columns = [i.get_text()[:-1] for i in table.find_all("th")]
    for i in table.find_all("td"):
        z = [y.get_text() for y in i.find_all("a")]
        v = [y.get_text() for y in i.find_all("i")]
        if len(v)==0:
            values.append(z)
        else:
            values.append(v)
    dic = {i:x for i,x in zip(columns,values)}
    return dic
```

```python
def vga(urls):
    years = list(range(2014,2019))
    dfs = []
    dics = []
    for i in urls:
        dics.append({})        
        tables = get_tables(i)
        n = urls.index(i)
        dics[n]["Year"] = years[n]
        dics[n]["Juri"] = to_dic(tables[0])
        dfs.append(pd.DataFrame(dics[n]))
        dics[n]["Fan"] = to_dic(tables[1])
        
        time.sleep(2)
    return dfs
```

```python
url_temp = "https://en.wikipedia.org/wiki/Category:%_video_games"
url_temp.replace("%", "2003")
```

```python
def get_list(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    return soup.find("div", {'id':'mw-pages'})
```

```python
temp = get_list(url_temp.replace("%", "2003"))
```

```python
[(i.get_text(), "https://en.wikipedia.org"+i["href"]) for i in temp.find("div").find_all("a")]
```

```python
def check_next(div):
    for i in div.find_all("a"):
        if "next page" in i.get_text():
            return True,i["href"]
    return False, ""
```

```python
games = []
for x in range(2003,2020):
    sopa = get_list(url_temp.replace("%",str(x)))
    games.append({})
    lst = [i.get_text() for i in sopa.find("div").find_all("a")]
    links = ["https://en.wikipedia.org"+i["href"] for i in sopa.find("div").find_all("a")]
    check = check_next(sopa)
    while check[0]==True:
        
        time.sleep(1)
        sopa = get_list("https://en.wikipedia.org/" + check[1])
        lst.extend([i.get_text() for i in sopa.find("div").find_all("a")])
        links.extend(["https://en.wikipedia.org"+i["href"] for i in sopa.find("div").find_all("a")])
        check = check_next(sopa)
    games[x-2003]["Year"] = x
    games[x-2003]["Games"] = lst
    games[x-2003]["Link"] = links
```

```python
def get_info_soup(teste):
    page = requests.get(teste)
    soup = BeautifulSoup(page.content, 'html.parser')
    return soup.find("table", class_=re.compile("infobox[ \w]*"))
```

```python
def get_info(info):
    dic = {}
    try:
        for i in info.find_all("tr")[2:]:
            try:
                dic[i.find("th").get_text()] = i.find("td").get_text()
            except:
                pass
    except:
        pass
    return dic
```

```python
def find_awards(game, df):
    awarded = []
    nominated = []
    for i in list(df.iterrows()):
        if game in i[1]["Winner"]:
            awarded.append(i[1]["Award"])
        if game in i[1]["Nominees"]:
            nominated.append(i[1]["Award"])
    if "Game of the Year" in awarded:
        game_year = 1
    else:
        game_year = 0
    print("Done")
    return awarded, nominated, game_year
```

```python
def nom(x):
    if len(x[1])==0:
        return 0
    elif len(x[1])==1:
        return x[1][0]
    return x[1]


def win(x):
    if len(x[0])==0:
        return 0
    elif len(x[0])==1:
        return x[0][0]
    return x[0]

def game_year(x):
    return x[2]
```

```python
df_games["Num_Awards"] = df_games["temp"].apply(lambda x: len(x[0]))
df_games["Num_Nominated"] = df_games["temp"].apply(lambda x: len(x[1]))
df_games["Awarded"] = df_games["temp"].apply(win)
df_games["nominated"] = df_games["temp"].apply(nom)
df_games["Won_Game_of_the_Year"] = df_games["temp"].apply(game_year)
```

```python
def pub(x):
    if x['Publisher(s)']:
        return x['Publisher(s)']
    elif x['Publisher']:
        return x['Publisher']
def dev(x):
    
    if type(x["Developer"]) == str:
        return x["Developer"]
    elif type(x["Developer(s)"])  == str:
        return x["Developer(s)"]
def plat(x):
    if x["Platform"]:
        return x["Platform"]
    elif x["Platform(s)"]:
        return x["Platform(s)"]
def rele(x):
    if x['Release']:
        return x['Release']
    elif x['Released']:
        return x['Released']
def gen(x):
    
    if type(x["Genre"]) != float:
        return x["Genre"]
    elif type(x["Genre(s)"]) != float:
        return x["Genre(s)"]

info["Publisher(s)"] = info.apply(pub, axis=1)
info["Developer(s)"] = info.apply(dev, axis=1)
info["Platform(s)"] = info.apply(plat, axis=1)
info["Release"] = info.apply(rele, axis=1)
info["Genre(s)"] = info.apply(gen, axis=1)
```

```python
def mes(x):
    meses = """January,February,March,April,May,June,July,August,September,October,November,December"""
    for i in meses.split(","):
        try:
            if i.lower() in x.lower() or i[:3].lower() in x.lower():
                return i
        except:
            pass
    return x
    
info["Release"] = info["Release"].apply(mes)
```
